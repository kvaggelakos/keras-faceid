```python
%matplotlib inline

from model import build_classifier
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns; sns.set()
from sklearn.cluster import KMeans
from sklearn import preprocessing
```

```python
data_path = './data'
```

```python
(clf, classess, X, Y) = build_classifier(data_path)
```

```python
X.shape
```

```python
classess
```

```python
X_norm = preprocessing.scale(X, axis=1)
```

```python
np.mean(X_norm, axis=1)
# np.std(X_norm, axis=1)
```

```python
kmeans = KMeans(n_clusters=5, random_state=0).fit(X_norm)
ymeans = kmeans.predict(X)
print(ymeans)
```

```python
pca = PCA(2)
projected = pca.fit_transform(X_norm)
plt.scatter(projected[:, 0], projected[:, 1],
            c=Y, edgecolor='none', alpha=1,
            cmap='viridis')
plt.colorbar()
```

```python
svc_clf = clf.predict(X_norm)
print(svc_clf)
pca = PCA(2)
projected = pca.fit_transform(X_norm)
plt.scatter(projected[:, 0], projected[:, 1],
            c=svc_clf, edgecolor='none', alpha=1,
            cmap='viridis')
plt.colorbar()
```
